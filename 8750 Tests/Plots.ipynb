```python
def plot_table(datas):
    
    data = datas['data']
    name = datas['name'] + '_table.pdf'
    fig = plt.figure(figsize=(10,5))
    
    columns = ('Field Min', 'Field Team', 'Forest Min', 'Forest Team', 'Building Min', 'Building Team')
    rows = ['%d %%' % x for x in (100, 98, 95, 90)]
    
    n_rows = len(rows)
    # Get some pastel shades for the colors
    colorsA = plt.cm.BuPu(np.linspace(0, 0.5, len(columns)))[0:n_rows][::-1]
    colorsB = plt.cm.GnBu(np.linspace(0, 0.5, len(columns)))[0:n_rows][::-1]
    colorsAB = zip(colorsA,colorsB)
    colors = colorsA
    
#     index = np.arange(len(columns)) + 0.1
#     bar_width = .8
    index = np.arange(len(columns)) + 0
    bar_width = 1
    
    # Initialize the vertical-offset for the stacked bar chart.
    y_offset = np.array([0.0] * len(columns))
    
    # Plot bars and create text labels for the table
    cell_text = []
    for row in range(n_rows):
#         y_offset = y_offset + data[row]
        plt.bar(index, data[row], bar_width, bottom=y_offset, color=colorsAB[row])
    #     cell_text.append(['%1.1f' % (x/1000.0) for x in y_offset])
        cell_text.append(['%1.1f' % (x) for x in data[row]])
    plt.ylim(datas['y_offset'],None)
    # Reverse colors and text labels to display the last value at the top.
#     colors = colors[::-1]
#     cell_text.reverse()
    
    # Add a table at the bottom of the axes
    the_table = plt.table(cellText=cell_text,
                          rowLabels=rows,
                          rowColours=colors,
                          colLabels=columns,
                          loc='bottom')
    
    # Adjust layout to make room for the table:
    plt.subplots_adjust(left=0.2, bottom=0.2)
    
    plt.ylabel("Time")
    plt.xticks([])
#     plt.title('Comparison of Progress over Method and Environment')
    plt.grid()
    
    fig.savefig(name, bbox_inches='tight')
    print 'Saved: ', name
    plt.show()
```
